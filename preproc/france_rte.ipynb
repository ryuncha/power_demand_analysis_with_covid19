```python
import glob
import os

import pandas as pd

from sqlalchemy import create_engine
import pymysql

import easydict

import database_info


from collections import Counter

pymysql.install_as_MySQLdb()
```

```python
engine, conn = database_info.getServerEngine()
```

```python
p = os.path.dirname(os.path.abspath(''))
```

```python
files = glob.glob(p+'/raw/france/*.xls')
```

```python
df = pd.DataFrame()

for file in files:
    tmp = pd.read_table(file, index_col=False, encoding='ISO-8859-1', error_bad_lines=False)
    df = pd.concat([df, tmp])
```

```python
print(df.columns.tolist())
```

```python
df = df.reset_index(drop=True)
```

```python
#      
translate = {'Primtre':'Country',
            'Nature':'Nature',
            'Date':'Date',
            'Heures':'Hours',
            'Consommation':'Consumption',
            'Prvision J-1':'Forecast D-1',
            'Prvision J':'Forecast D-day',
            'Fioul':'Fuel oil',
            'Charbon':'Coal',
             'Gaz':'Gas',
             'Nuclaire':'Nuclear',
             'Eolien':'Wind',
             'Solaire':'Solar',
             'Hydraulique':'Hydraulic',
             'Pompage':'Pumping',
             'Bionergies':'Bionergies',
             'Ech. physiques':'Ech. physiques',
             'Taux de Co2':'Co2 rate',
             'Ech. comm. Angleterre':'Ech. comm. England',
             'Ech. comm. Espagne':'Ech. comm. Spain',
             'Ech. comm. Italie':'Ech. comm. Italy',
             'Ech. comm. Suisse':'Ech. comm. Swiss',
             'Ech. comm. Allemagne-Belgique':'Ech. comm. Germany-Belgium',
             'Fioul - TAC':'Fuel oil - TAC',
             'Fioul - Autres':'Fuel oil - Others',
             'Gaz - TAC':'Gaz - TAC',
             'Gaz - Cogn.':'Gas - Cogen.',
             'Gaz - CCG':'Gas - CCG',
             'Gaz - Autres':'Gas - Others'}
```

```python
ndf = df.rename(columns=translate)
```

```python
# ndf  null row 
ndf = ndf[ndf.Country=='France']
```

```python
date_hours = ndf.Date.apply(lambda x:str(x)+' ') + +ndf.Hours.apply(lambda x:str(x))
```
